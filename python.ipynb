***4. Continue statement: used inside loops to skip the rest of the code inside the loop for the current iteration and proceed to the next iteration***

```python
# Example of continue statement:
for i in range (1, 11):
   if i % 2 == 0:
     print("Before Continue")
     continue  # skips current iteration of displaying even numbers
     print("After Continue")
   print("Inside loop")
   print(i, end=" ")   #print odd numbers between 1 to 10
```

**5. Break statment: Used to exit the loop permanently based on certain conditions.**


```python
# example of break statement
for i in range (1, 11):
  if i > 5:
    print("Before break")
    break  #  stop the loop when i becomes greater than 5
    print("After break")  # This will never get executed as it after break starts. (Dead code.)
  print(i, end=" ")  # print numbers 1 to 5 seperated by space..
```

```python
# Write a python function Print_even(x,y), to print all the even numbers between x and y
def print_even(x,y):
  for i in range(x,y+1):
    if i%2==0:
      print(i, end=" ")

print_even(10,30)
print("\n")
print_even(51,70)
```

```python
# Write a python function Print_prime(x,y), to print all the Prime numbers between x and y
def print_prime(x,y):
  for i in range(x,y+1):
    if i>1:
      for j in range(2,i):
        if i%j==0:
          break
      else:
        print(i, end=" ")

print_prime(1,20)
```

**4. Find words starting with letter**

write a function word_starting_with(words, letter) that:
      Take a list of string and a charater.
      return the list of words that start with the given charater

    Input : (["banana", "apple", "apricot"], "a")  
    output : ["apple", "apricot"]

```python
def word_starting_with(words, letter):
   starting_with_letter = []
   for word in words:
     if word.startswith(letter):
       starting_with_letter.append(word)

   return starting_with_letter

word_starting_with(words=["apple", "apricot", "banana"], letter="a")
```

```python
def word_starting_with(words, letter):
  new_list = []
  for item in words:
    if item[0] == letter:
      new_list.append(item)

  return new_list

word_starting_with(words=["apple", "apricot", "banana"], letter="a")
```

```python
def word_starting_with(words, letter):
   return [item for item in words if item[0]==letter]

word_starting_with(words=["apple", "apricot", "banana"], letter="a")
```

**5. Remove Duplciates and sort**

write a function unique_sorted(numbers) that:

    take a list of intergers.
    Returns a sorted list with duplicate removed.

    Input = [1,2,3,4,2,5,4]
    output = [1,2,3,4,5]

```python
def unique_sorted(numbers):
  sorted_numbers = set(numbers)
  return sorted(sorted_numbers)

unique_sorted(numbers = [1,2,3,4,3,5,4])
```

```python
def unique_sorted(numbers):
  sorted_numbers = set(numbers)
  new_list = list(sorted_numbers)
  new_list.sort()
  return new_list

unique_sorted(numbers = [1,2,3,4,3,5,4])
```

```python
def unique_sorted(numbers):
  for item in numbers:
    if numbers.count(item) >1:
      numbers.remove(item)

  numbers.sort()
  return numbers

unique_sorted(numbers = [1,2,3,4,3,5,4])
```

```python
my_list = [1, 2, 3, 4, 2, 5, 4]
seen = set()
duplicates = []

for item in my_list:
  if item in seen:
    duplicates.append(item)
  else:
    seen.add(item)

print("Original list:", my_list)
print("Duplicate numbers:", duplicates)
```

**6. Find longest word**

write a function longest_word(words) that:

    take a list of strings.
    Return the longest word. if they are ties return the first one.

    Input = ["apple", "banana", "cherry"]
    output = banana

```python
def longest_word(words):
  longest_word = ""
  for item in words:
    if len(item) > len(longest_word):
      longest_word = item

  return longest_word

longest_word(words=["apple", "apricot", "banana"])
```

      **MAP Functions**
*   The map function in python applies a given fuction to each item of an iterable (such as a list, tuple, or set) and retrunsa new iterator that yielss the results.
*   It takes two arguments: the function to apply and the iterable to apply it to.
*   The function provided to map can be a build-in function, a lambda function, or a user-defined function.
*   Map is commonly used to apply a transformation or operation to all elements of a collection without the need for explicit loops.
*   It offers a more concise and efficient alternative to using a for loop with a list comprehension.
















```python
# Example of map functions
def square(x):
  """function to have a square of the numbers"""
  return x**2

#Applying square function to each element in the list using map
numbers=[1, 2, 3, 4]
squared_numbers = map(square, numbers)

# Converting the map object to a list
squared_numbers_list = list(squared_numbers)
print("Squared numbers:" , squared_numbers_list)

# In this example, the map function is used to apply the square function to each element of the numbers list, remaining in a new list containing
```
