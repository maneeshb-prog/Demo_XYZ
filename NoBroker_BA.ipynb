```python
import pandas as pd
```

```python
hprices = pd.read_csv('homeprices.csv')
hprices
```

```python
# Descriptive Analysis :- Describing the past

hprices.describe()
```

```python
import matplotlib.pyplot as plt

plt.scatter(hprices.area, hprices.price)
plt.xlabel('Area')
plt.ylabel('price')
plt.show()
```

```python
# Creating a DataFrame/table for independent variable. And it will have 2 brackets...

x=hprices[['area']]
x
```

```python
# Creating a Series or a column. its a dependent variable on Data frame and Series/column for target value...

y=hprices['price']
y
```

```python
# Predective Analysis :-- Forecasting techniques and Statistical models to find, what is going to have in future...

from sklearn.linear_model import LinearRegression

model = LinearRegression()
model.fit(x,y)
```

```python
model.predict([[3300]])
```

```python
import matplotlib.pyplot as plt

plt.scatter(hprices.area, hprices.price)
plt.xlabel('Area')
plt.ylabel('Price')
plt.title('Area vs Price')
plt.show()
```

```python
import pandas as pd

# Create a DataFrame for prediction to match the format used during training
prediction_data = pd.DataFrame({'area': [3400]})
predicted_price = model.predict(prediction_data)
print(f"The predicted price for a 3400 sq.ft home is: ${predicted_price[0]:,.2f}")
```

```python
import pandas as pd

areas_to_predict = pd.DataFrame({
    'area': [2200, 7200, 5200, 3700]
})

predicted_prices = model.predict(areas_to_predict)

prediction_df = pd.DataFrame({
    'Area (sqft)': areas_to_predict['area'],
    'Predicted Price': predicted_prices
})

display(prediction_df.style.format({'Predicted Price': '${:,.2f}'}))
```
